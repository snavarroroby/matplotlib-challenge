## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
cleaned_df

df = cleaned_df.groupby('Mouse ID')
final_timepoint = df['Timepoint'].max()
final_timepoint

final_timepoint_df = pd.DataFrame(final_timepoint)
final_timepoint_df

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

merge_df = pd.merge(cleaned_df, final_timepoint_df, on=['Mouse ID','Timepoint'], how='inner')
merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
         
# Create empty list to fill with tumor vol data (for plotting)
tum_vol = []

for drug in drugs:
    
    drug_loc = merge_df.loc[merge_df['Drug Regimen']== drug]
    
    tum_vol.append(drug_loc['Tumor Volume (mm3)'])

      
       
        
# Calculate the IQR and upper/lower bounds for each drug to quantitatively determine if there are any potential outliers. 

#Calculating upper/lower bounds for Capomulin
quartiles_capomulin = tum_vol[0].quantile([.25,.5,.75])
lowerq_capomulin = quartiles_capomulin[.25]
lowerq_capomulin
upperq_capomulin = quartiles_capomulin [.75]
upperq_capomulin

iqr_capomulin = upperq_capomulin - lowerq_capomulin
    
low_bound_capomulin = lowerq_capomulin - (1.5*iqr_capomulin)
upper_bound_capomulin = upperq_capomulin + (1.5*iqr_capomulin)

print('------------------------------')
print(f"Values below {low_bound_capomulin} could be outliers for capomulin.")
print(f"Values above {upper_bound_capomulin} could be outliers for capomulin.")


#Calculating upper/lower bounds for Ramicane
quartiles_ramicane = tum_vol[1].quantile([.25,.5,.75])
lowerq_ramicane = quartiles_ramicane[.25]
upperq_ramicane = quartiles_ramicane [.75]

iqr_ramicane = upperq_ramicane - lowerq_ramicane
low_bound_ramicane = lowerq_ramicane - (1.5*iqr_ramicane)
upper_bound_ramicane = upperq_ramicane + (1.5*iqr_ramicane)

print('------------------------------')
print(f"Values below {low_bound_ramicane} could be outliers for ramicane.")
print(f"Values above {upper_bound_ramicane} could be outliers for ramicane.")


#Calculating upper/lower bounds for Infubinol
quartiles_infubinol = tum_vol[2].quantile([.25,.5,.75])
lowerq_infubinol = quartiles_infubinol[.25]
upperq_infubinol = quartiles_infubinol[.75]

iqr_infubinol = upperq_infubinol - lowerq_infubinol
low_bound_infubinol = lowerq_infubinol - (1.5*iqr_infubinol)
upper_bound_infubinol = upperq_infubinol + (1.5*iqr_infubinol)

print('------------------------------')
print(f"Values below {low_bound_infubinol} could be outliers for infubinol.")
print(f"Values above {upper_bound_infubinol} could be outliers for infubinol.")

#Calculating upper/lower bounds for Ceftamin
quartiles_ceftamin = tum_vol[3].quantile([.25,.5,.75])
lowerq_ceftamin = quartiles_ceftamin[.25]
upperq_ceftamin = quartiles_ceftamin[.75]

iqr_ceftamin = upperq_ceftamin - lowerq_ceftamin
low_bound_ceftamin = lowerq_ceftamin - (1.5*iqr_ceftamin)
upper_bound_ceftamin = upperq_ceftamin + (1.5*iqr_ceftamin)

print('------------------------------')
print(f"Values below {low_bound_ceftamin} could be outliers for ceftamin.")
print(f"Values above {upper_bound_ceftamin} could be outliers for ceftamin.")

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
#plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.annotate(line_eq, (19,37), fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Relationship of Mouse Weight vs. Average Tumor Volume for Mice Treated with Capomulin')
```
